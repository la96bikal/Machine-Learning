```python
import pandas as pd
from itertools import combinations
import time
```

```python
data = pd.read_csv('SuperCenterDataNew.csv', header= None)
```

<h1> Getting 5 % of the data

```python
dataset = data.head(int(0.05 * len(data)))
```

<h2> Dataset to a list </h2>
This Function converts the pandas data frame in to a list of the rows and drops all the na values 

```python
def dataset_to_list(dataset):
    dataset_into_list = []
    for index,rows in dataset.iterrows():
        rows = rows.dropna()
        dataset_into_list.append(rows.tolist())
    return dataset_into_list[:]
```

<h2> Pruning the data list </h2>
This function prunes the data list as we iterate through different item sizes.
Reduction of the rows improves efficiency as the combinations of the rows will decrease with the decrement in row size.
It takes in the original datalist and the frequent list.
Frequent list should only contain single elements and tuple values

```python
def normalize_data(data_list, frequent_list):
    frequent_list = set(frequent_list)
    new_dataset = []
    for row in data_list:
        row = set(row)
        if row.intersection(frequent_list) != set():
            new_dataset.append(list(row))
    return new_dataset[:]
```

<h2> Tuples to elements </h2>
This function is used to change the tuple values to single elements and remove all the replicates using set properties.

```python
def tuple_to_elements(freq_item_list):
    elements = []
    for i in freq_item_list:
        elements.extend([e for e in i])
    return list(set(elements))
```

<h2> This function is the frequent item set miner </h2>

```python
def frequent_item_set(datalist, absolute_support):
    start_time = time.time()
    frequent_items_dict = {}  
    
    i = 1
    
    while True:
        if i == 1:
            
            item_dict = {}
            
            print("Item size ::", i, " Absolute support ::", absolute_support)
                 
            for rows in datalist:
                for row_element in rows:
                    try:
                        item_dict[row_element] += 1
                    except KeyError:
                        item_dict[row_element] = 1
            
            frequent_items =  [i for i in item_dict.keys() if item_dict[i] >= absolute_support]
            frequent_items_dict[i] = frequent_items
            datalist = normalize_data(datalist, frequent_items)
            print("Datalist size after normalization ", len(datalist))
            print("Frequent items size", len(frequent_items))
            
            print(frequent_items)
            
            print()
            
        else:
            if len(frequent_items_dict[i-1]) < 1:
                break;
                
            print("Item size ::", i, " Absolute support ::", absolute_support)
            frequent_pairs = {}
            item_dict = {}
            
            unit_percent = len(datalist) // 100            
            counter = 0
            percent_progress = 0
            for row in datalist:  
                if counter % unit_percent == 0:
                    percent_progress += 1
                    print("%d% % done. \r"%percent_progress, end="")
                
                if (len(row) < i):
                    continue
                
                row_combinations = combinations(row, i)
                for c in row_combinations:
                        try:                            
                            item_dict[c] += 1                            
                        except:                           
                            item_dict[c] = 1
                counter += 1
            
            frequent =  [i for i in item_dict.keys() if item_dict[i] >= absolute_support]
            
            frequent_items_dict[i] = frequent
            frequent_indiv_elem = tuple_to_elements(frequent)
            datalist = normalize_data(datalist, frequent_indiv_elem)
            print("Datalist size after normalization ", len(datalist))
            print("Frequent items size", len(frequent))
            
            print(frequent)
            
            print()
            
            
            
        
        i += 1
          
        
        print("Time elapsed :: ", time.time() - start_time)
    return frequent_items_dict
```
