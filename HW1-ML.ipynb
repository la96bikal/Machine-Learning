<h1> Homework 01 - Simple Linear Regression </h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

<h2> Reading the dataset using pandas </h2>

```python
data = pd.read_csv('Homework1Datasaet.csv')
data
```

<h2> Creating Dummy variables with drop_first set to true. </h2>
<h2> Crust properties have prefix 'Category_Crust' and Filling properties have prefix 'Category_Filling' </h2>

```python
crust_columns = ['Crust Shape', 'Crust Size', 'Crust Shade']
filling_columns = ['Filling Size', 'Filling Shade ']

crust_dummies = pd.get_dummies(data, drop_first = True, prefix = 'Category_Crust', columns = crust_columns)
filling_dummies = pd.get_dummies(data, drop_first = True, prefix = 'Category_Filling', columns = filling_columns)

crust_dummies = crust_dummies.drop(columns = filling_columns)
filling_dummies = filling_dummies.drop(columns = crust_columns)

final_dataset = crust_dummies.merge(filling_dummies, how = 'inner')
```

```python
final_dataset
```

<h2> Splitting independent and dependent variables as x_training and y_training where x_training holds all the features and y_training holds the classes </h2>

```python
y_training = final_dataset['Class']
y_training = y_training.replace(['pos', 'neg'], ['1', '0'])
x_training = final_dataset.drop(columns = ['Example', 'Class'])
```

<h2> Using Scikit-learn to fit a linear regression model to the dataset </h2>

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg = LinearRegression().fit(x_training, y_training)
y_predicted = linear_reg.predict(x_training)
y_predicted
```

```python
y_predicted = [1 if x >= 0.5 else 0 for x in y_predicted]
actual_vs_predicted  = pd.DataFrame()
actual_vs_predicted['actual_pos'] = y_training
actual_vs_predicted['predicted'] = y_predicted
```

```python
actual_vs_predicted
```

We can see that the linear regression model has an accuracy of 100 % accuracy on the training set
