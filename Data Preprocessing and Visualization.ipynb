```python
import pandas as pd
```

```python
data = pd.read_csv('uncleandata.csv')
```

```python
data.head()
```

<h1> Total row Duplicates </h1>

```python
data.duplicated().sum()
```

<h1> Removing duplicates </h1>

```python
data.drop_duplicates()
data.head()
```

```python
data.isna().sum()
```

<h1> Filling missing values with mean </h1>

```python
data = data.fillna(data.mean())
```

<h1> Dropping the column 'is_smoker' as it has the largest number of missing values </h1>

```python
data = data.drop(columns = ['is_smoker'])
```

```python
data.isna().sum()
```

```python
data['left'] = data.left.eq('yes').mul(1)
```

```python
data.head()
```

<h1> Distribution of all the features </h1>

<i> Plotting histogram of quantitative features and countplots for categorical features

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
categorical_features = ['left', 'department', 'salary'] 
quantitative_features = list(set(data.columns) - set(categorical_features))

count = 1
fig = plt.figure(figsize=(13,13))
for i in quantitative_features:
    plt.subplot(3,3, count)
    sns.distplot(data[i])
    count += 1
```

```python
fig = plt.figure(figsize=(10,10))
count = 1

for i in categorical_features:
    plt.subplot(3,1, count)
    sns.countplot(data[i])
    count += 1
```

```python
data.to_csv('CleanData.csv')
```
